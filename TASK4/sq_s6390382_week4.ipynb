```python
#make lists
X = [1,2,3,10]
Y = [10, 20, 30, 40]
#define z as zip of x and y
Z = zip(X,Y)
X_Y = [] #empty list

for x, y in Z:
    X_Y.append(x+y) #add x+y values to empty list

print(f"Result of adding x and y (zip):", X_Y)
```

```python
X = [1,2,3,10]
Y = [10, 20, 30, 40]


print(f"Result of adding x and y (zip one liner):", [(x+y) for x, y in zip(X,Y)])
```

```python
X = [1,2,3,10]
Y = [10, 20, 30, 40]
def add(a,b):  # make funtion that calculates x = y
    return a+b
added = map(add, X, Y) # use map t
print(f"Result of adding x and y (map):", list(added))
```

```python
from matplotlib import pyplot as plt

lengths = np.random.normal(185, 8, 100000) # Make a numpy array of normal distribution with mean 185, and standard deviation of 8, size 100000 

mean = np.mean(lengths) # Take the mean of the array lengths
sd = np.std(lengths) # Take the standard deviation of the array
print(mean, sd) # Print the mean and standard deviation


plt.hist(lengths, 40, label = "Hist") # Plot the histogram of lengths with 40 bins
plt.axvline(mean, marker= 'o', color = 'blue', markersize = 3, label = "mean")
plt.axvspan(mean-sd, mean+sd, color = 'red', alpha = 0.2, label = "standard dev")
plt.legend()
plt.show # Show the histogram


# Check if the difference between the calculated value and the input value is smaller than 0.5 so we know if they are similar or not
if  abs(mean-185) < 0.5 and abs(sd-8) < 0.5:
    print("They are similar")
else:
    print("They are not similar")


from scipy.stats import norm # Import norm from scipy.stats library

cdf_value = norm.cdf(190, loc=mean, scale=sd) # Calculate probability that value = 185
print(f"CDF {cdf_value}")

cdf = norm.cdf(mean+sd, mean, sd) - norm.cdf(mean-sd, mean, sd)  # Calculate probability that value is between one sd away from mean on both sides
print(f"CDF {cdf}")
```

```python
import numpy as np # Imports numpy library as alias np

def monte_carlo_int(func, a, b, N): # Define function with variables func, a (lower lim), b (upper lim), N (number of points)
    """ Monte calro integration between a and b"""
    
    x = np.random.uniform(a, b, N) # Generate N amount of x values uniform between a and b
    y = eval(func) # Evaluates the function put in when called upon the udf

    integ = (b-a) * np.mean(y) # Calculate the integral using that the (sum of f(x))/n is equal to the mean of all f(x)
    return integ

func = input("Enter function you want the integral of:") # Define func as input
a = float(input("Enter the lower bound:")) # Define a (lower bound in integral) as input
b = float(input("Enter the upper bound:")) # Define b (upper bound in integral) as input
N = int(input("Enter the number of samples to use in integral (like 1000):")) # Define N (number of samples) as input
float(monte_carlo_int(func, a, b, N)) # Should be around 1/2 since integral of x between 0 and 1 is 1/2, float to get rid of np.float64 infront
```

```python
# Import libraries with aliases
import numpy as np
import matplotlib.pyplot as plt

# Define arrays as given
x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

# combine into the complex numbers
z = x +1j*y

# angle of rotation
deg = float(input("Give degrees of wanted rotation:"))
rad = np.deg2rad(deg)

#define rotor
rotor = np.exp(1j * rad) # Factor to rotate with

rotatedz = rotor * z # calculate new square after rotation
rotatedx = rotatedz.real # Take the real part of z (complex) to get the new x value
rotatedy = rotatedz.imag # Take the imaginary part of z (complex) to get the new y value

#plot original square
plt.plot(x,y, color = 'red', label = 'Original square')
plt.plot(rotatedx,rotatedy, color = 'blue', label = 'Rotated square')
plt.legend()
plt.show()
```

```python
import numpy as np

def sum_arrays(*args):
    """
    Calculates the sum of any amount of arrays
    """
    s = np.zeros_like(args[0]) # make a new array that is formatted like the first input array but has values all zero
    
    for array in args:
        s += array # adds the array to the original array of all zeros. If done for all arrays the answer will be the total sum of all arrays
    return s

#define test arrays
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.array([7,8,9])

#print sums of test arrays using function
print(sum_arrays(a))
print(sum_arrays(a,b))
print(sum_arrays(a,b,c))
```

```python
def funcfilter(func, seq):
    """
    Filters out non positive and zero numbers from seq
    """
    array = np.asarray(seq) # define the array of seq as an array (that supports all number types for seq)
    filtered = array[array>0] # define filtered array as only the items from array that are larger than 0 (so 0 also excluded)
    return func(filtered) # apply the function (put in in the udf when called upon) to the filtered array

# Test with provided array and function
s = np.array([-100,-10,0,10,100])
f = np.log10
result = funcfilter(f, s)
print("Result log10:", result)

# Test with single number and function
s = 100
f = np.log10
result = funcfilter(f, s)
print("Result log10:", result)
```
