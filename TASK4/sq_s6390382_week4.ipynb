```python
from matplotlib import pyplot as plt

lengths = np.random.normal(185, 8, 100000) # Make a numpy array of normal distribution with mean 185, and standard deviation of 8, size 100000 

mean = np.mean(lengths) # Take the mean of the array lengths
sd = np.std(lengths) # Take the standard deviation of the array
print(mean, sd) # Print the mean and standard deviation


plt.hist(lengths, 40, label = "Hist") # Plot the histogram of lengths with 40 bins
plt.axvline(mean, marker= 'o', color = 'blue', markersize = 3, label = "mean")
plt.axvspan(mean-sd, mean+sd, color = 'red', alpha = 0.2, label = "standard dev")
plt.legend()
plt.show # Show the histogram


# Check if the difference between the calculated value and the input value is smaller than 0.5 so we know if they are similar or not
if  abs(mean-185) < 0.5 and abs(sd-8) < 0.5:
    print("They are similar")
else:
    print("They are not similar")


from scipy.stats import norm # Import norm from scipy.stats library

cdf_value = norm.cdf(190, loc=mean, scale=sd) # Calculate probability that value = 185
print(f"CDF {cdf_value}")

cdf = norm.cdf(mean+sd, mean, sd) - norm.cdf(mean-sd, mean, sd)  # Calculate probability that value is between one sd away from mean on both sides
print(f"CDF {cdf}")
```

```python
import numpy as np # Imports numpy library as alias np

def monte_carlo_int(func, a, b, N): # Define function with variables func, a (lower lim), b (upper lim), N (number of points)
    """ Monte calro integration between a and b"""
    
    x = np.random.uniform(a, b, N) # Generate N amount of x values uniform between a and b
    y = eval(func) # Evaluates the function put in when called upon the udf

    integ = (b-a) * np.mean(y) # Calculate the integral using that the (sum of f(x))/n is equal to the mean of all f(x)
    return integ

func = input("Enter function you want the integral of:") # Define func as input
a = float(input("Enter the lower bound:")) # Define a (lower bound in integral) as input
b = float(input("Enter the upper bound:")) # Define b (upper bound in integral) as input
N = int(input("Enter the number of samples to use in integral (like 1000):")) # Define N (number of samples) as input
float(monte_carlo_int(func, a, b, N)) # Should be around 1/2 since integral of x between 0 and 1 is 1/2, float to get rid of np.float64 infront
```

```python
import numpy as np

def sum_arrays(*args):
    """
    Calculates the sum of any amount of arrays
    """
    s = np.zeros_like(args[0]) # make a new array that is formatted like the first input array but has values all zero
    
    for array in args:
        s += array # adds the array to the original array of all zeros. If done for all arrays the answer will be the total sum of all arrays
    return s

#define test arrays
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.array([7,8,9])

#print sums of test arrays using function
print(sum_arrays(a))
print(sum_arrays(a,b))
print(sum_arrays(a,b,c))
```

```python
def funcfilter(func, seq):
    """
    Filters out non positive and zero numbers from seq
    """
    array = np.asarray(seq) # define the array of seq as an array (that supports all number types for seq)
    filtered = array[array>0] # define filtered array as only the items from array that are larger than 0 (so 0 also excluded)
    return func(filtered) # apply the function (put in in the udf when called upon) to the filtered array

# Test with provided array and function
s = np.array([-100,-10,0,10,100])
f = np.log10
result = funcfilter(f, s)
print("Result log10:", result)

# Test with single number and function
s = 100
f = np.log10
result = funcfilter(f, s)
print("Result log10:", result)
```

```python
# Import libraries with alias
from math import sqrt
import numpy as np
import scipy as sp
from astropy.cosmology import Planck18

# Define constants as given in question
H0 = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0

# Define function that calculates the inside of the integral witht the given value for z
def tL(z):
    H = H0 * sqrt(Om0 * ((1+z)**3) + Ode0)
    F = (1 / ( (1+z) * H))
    return F

# Give values for in the function
a = 0
z = 2

# Calculate the integral of the function using the values and turn it into a numpy array
solution, error = np.array(sp.integrate.quad(tL, a, z))
conversion  = (3.1710e-17 * 3.0856776e19) # Conversion factor to go from seconds to gigayear
result = solution * conversion
result_error = conversion * error

# Print the solution
print("Solution is:", result, "+-", result_error, "Gyr")
```

```python
# Import libs
import matplotlib.pyplot as plt # for plotting
import numpy as np # for arrays and calculations
from scipy.stats import norm #for pdf


# Make arrays with normal distribution, Mean, standard dev, amount of data points
X = np.random.normal(0,1,100000)
Y = np.random.normal(2,1,100000)

#calculate mean and standard deviation of X and Y with numpy
meanx = np.mean(X)
meany = np.mean(Y)
sdx = np.std(X)
sdy = np.std(Y)

# Calculate probability that value is between one sd away from mean on both sides
value = np.linspace(-5,7,1000) # interval and amount of data points
pdfX = norm.pdf(value, meanx, sdx) # calculates the probability density for each point in value with mean and standard dev previously calculated
pdfY = norm.pdf(value, meany, sdy)

# Plot the histograms of X and Y with 30 bins
plt.hist(X, 30, density = True, alpha = 0.3, label = 'x', color = 'orchid', edgecolor = 'black')
# arrat to plot, amount of bins, needed for plotting pdf later, transparency, label, color of graph, edgecolor of graph
plt.hist(Y, 30, density = True, alpha = 0.3, label = 'y', color = 'blue', edgecolor = 'black')

# Plot the pdf of X and Y 
plt.plot(value, pdfX, label = 'PDF X', color = 'purple', linewidth = 2, linestyle = ':')
# x axis, y axis, lavel, color of line, line thickness
plt.plot(value, pdfY, label = 'PDF Y', color = 'blue', linewidth = 2, linestyle = '--')

#Change tick rate on x axis
plt.xticks(np.arange(-5,7,1)) #start, stop, stepsize

# Add arrows pointing to pdf lines
plt.annotate(r"$\mathcal{N}$(0,1)", xy=(-2,0.155), xytext=(-5, 0.15), arrowprops=dict(arrowstyle="simple", color='purple', lw=3,))
#name nest to arrow, point where arrow points, location of text/starting of arrow, arrow looks
plt.annotate(r"$\mathcal{N}$(2,1)", xy=(4,0.155), xytext=(5.8, 0.15), arrowprops=dict(arrowstyle="simple", color='blue', lw=3,))

# Add title and labels
plt.title(' 1D histogram + pdf')
plt.xlabel('Values')
plt.ylabel('Probability Density')

# add grid with specific line style only on y axis and add legend including all labels previously mentioned
plt.grid(linestyle = '--', axis = 'y')
plt.legend()
plt.show


```

```python
# Compute the 2D histogram
H, xedges, yedges = np.histogram2d(X, Y, bins=30, density=True)
#edges are the edge of each bin, so xedges is on the x axis , y egdes on y axis

# Plot the histogram
plt.figure() #makes sure its a new graph
plt.pcolormesh(xedges, yedges, H.T, cmap="Blues") # coloured grid
plt.colorbar(label="Probability Density") # side bar

# Compute bin centers for contours
xcenters = 0.5 * (xedges[:-1] + xedges[1:]) # takes center of bins
ycenters = 0.5 * (yedges[:-1] + yedges[1:])
Xgrid, Ygrid = np.meshgrid(xcenters, ycenters) # makes the grids for x and y

# Use histogram values for contours
Z = H.T # makes sure bins are not rotated by transposing bins

# Contour levels
Z_flat = Z.ravel() # flattens multidimensional array into a single row
Z_sort = np.sort(Z_flat)[::-1] # flips array
Z_cumsum = np.cumsum(Z_sort) # cumulative sum
Z_cumsum /= Z_cumsum[-1]  # divide by last aka normalise

# Define the contour levels corresponding to 1σ, 2σ, and 3σ
levels = [Z_sort[np.searchsorted(Z_cumsum, p)] for p in [0.68, 0.95, 0.997]] # search sorted looks for the indices that correspond to the elements that shoudl be inserted to maintain order
levels_use = levels[::-1] # flip array around (back to original)

# Plot contours
cs = plt.contour(Xgrid, Ygrid, Z, levels=levels_use,
                 colors="red", linewidths=1.5) # define countour with previous grids and make pretty

# Add labels to contours (this will add labels like "1σ", "2σ", "3σ" on the plot)
plt.clabel(cs, inline=True, fontsize=12, fmt={levels_use[0]:"3σ", levels_use[1]:"2σ", levels_use[2]:"1σ"})

# Add grid, labels, and title adn ticks
#Change tick rate on x axis
plt.xticks(np.arange(-4,5,2))
plt.yticks(np.arange(-2,7,1))

plt.grid(linestyle="--", linewidth=0.7)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2D Histogram + Contour")

# Show the plot
plt.show()
```
