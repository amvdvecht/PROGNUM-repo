| File organisation commands      |                   What it does |
| :----------- | ------------------------------------------------: |
| mkdir FILES      |          Creates a new directory called `FILES` in your current directory |
| rmdir FILES      | Removes empty directory `FILES`          |
| touch a.txt         |    Creates a new empty file called `a.txt`        |
| cp a.txt b.txt      |       Copies file `a.txt` to `b.txt`              |
| cp a.txt .         | Copies file `a.txt` into current directory (`.` = current dir) |
| cp -r * FILES |         Copies all contents recursively from current directory to `FILES` |
| mv b.txt c.txt       |     Renames or moves file `b.txt` to `c.txt`      |
| mv a.py bn.py        | Renames file `a.py` to `b.py`             |
| mv a.py Python/       | Moves file a to directory `Python`               |
| rm file           | Removes file `file`                      |
| rm -i file              | Prompts before removing file                   |
| rm -f file              | Removes file without prompt (dangerous)        |
| rm -r directory         | Removes directory and all contents             |
| ln -s target linkname | Creates a symbolic link (shortcut) to a file                     |

| File viewing | Description                                       |
| :--------------- | ------------------------------------------------- |

| tail -100 file   | Shows last 100 lines of a file (default 10)       |
| more report.doc  | Pager to view file one page at a time             |

| file report.doc  | Shows the type of a file                          |




| Redirection and pipes         | Action                                                       |
| :------------------------- | ------------------------------------------------------------ |
| >                          | Redirects output to a file (overwrites if exists)            |
| >>                         | Appends output to a file                                     |
| <                          | Uses a file as input for a command                           |
| | (pipe)                   | Pipes output of one command as input to another              |
| ls > filenames.txt         | Writes `ls` output to file `filenames.txt`                   |
| ls >> filenames.txt        | Appends `ls` output to `filenames.txt`                       |
| who > temp                 | Saves list of users to file `temp`                           |
| sort < temp > users_sorted | Sorts contents of `temp` and writes output to `users_sorted` |
| who | sort                 | Sorts the output of `who` command using a pipe               |
| who | sort > users_sorted  | Same as above but saves sorted output to `users_sorted`      |


| Movement/scroll Options   | Action                                        |
| :------- | --------------------------------------------- |
| q        | Stops less/more pager                         |
| spacebar | Moves down one page                           |
| g        | Goes to beginning of document                 |
| G        | Goes to end of document                       |
| /hello   | Searches forward for "hello"                  |
| ?hello   | Searches backward for "hello"                 |
| n        | Repeats previous search in forward direction  |
| N        | Repeats previous search in backward direction |

| Wildcard / Symbol | Meaning                                                                   |
| :---------------- | ------------------------------------------------------------------------- |
| *                 | Matches zero or more characters (not slash / or first dot of hidden file) |
| ?                 | Matches exactly one character (not slash / or first dot of hidden file)   |
| [abz] or {a,b,z}  | Matches any one character listed (a, b, or z)                             |
| [a-z]             | Matches any character from a to z                                         |
| [0-9]             | Matches any digit from 0 to 9                                             |
