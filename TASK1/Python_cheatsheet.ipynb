| type of comment         |                                Description |
| :-------------- | -----------------------------------------: |
| inline         |                explains specific line of code which it is after |
| Block         |                Explains full block of following code |
| Docstrings       |                Explains full block of following code |

| Command         |                                Description |
| :-------------- | -----------------------------------------: |
| print()         |                Prints output to the screen |
| input()         |           Gets user input (returns string) |
| type()          |             Returns the type of a variable |
| del             |                         Deletes a variable |
| import          |                           Imports a module |
| from … import … |      Imports specific object from a module |
| from … import * |      Imports all objects from module |
| keyword.kwlist  |                  List of reserved keywords |
| .value          |  Accesses the value attribute of an object |
| .real           |      Returns real part of a complex number |
| .imag           | Returns imaginary part of a complex number |

| Print optional Parameters |                         Description |
| :-------- | ----------------------------------: |
| sep       | Separator between printed arguments |
| end       | String appended at end (default \n) |
| file      |                  Output destination |
| flush     |     Forces immediate output if True |


| Types of "variables"  |                               Description |
| :-------- | ----------------------------------------: |
| int()     |               Converts value to integer   |
| float()   | Converts value to floating point number   |
| complex() |        Converts value to complex number   |
| str()     |                Converts value to string   |
| eval()    | Evaluates a string as a Python expression |

| String Operator |          Description |
| :------- | -------------------: |
| +        | Concatenates strings |
| *        |     Repeats a string |
| .upper() | Converts string to uppercase |
| .lower() | Converts string to lowercase |
| .split() |    Splits string into a list |

| String Formatting    |                 Meaning |
| :--------- | --------------------------------: |
| f"..."     |          Formatted string literal |
| r"..."     | Raw string (no escape processing) |
| {}         |          Placeholder in f-strings |
| {variable} |          Placeholder for variable |
| {var:10f}  |      Field width 10, float format |
| {var:6.3f} |              Width 6, precision 3 |
| {var:e}    |               Scientific notation |
| {var:g}    |                    General format |
| {var:%}    |                 Percentage format |



| Operator |         Description |
| :------- | ------------------: |
| +        |            Addition |
| -        |         Subtraction |
| *        |      Multiplication |
| /        |            Division |
| %        | Modulus (remainder) |
| **       |      Exponentiation |
| //       |      Floor division |

| Operator |  Equivalent To |
| :------- | -------------: |
| =        |      x = value |
| +=       |  x = x + value |
| -=       |  x = x - value |
| *=       |  x = x * value |
| /=       |  x = x / value |
| %=       |  x = x % value |
| //=      | x = x // value |
| **=      | x = x ** value |

| Symbol |                                    Meaning |
| :----- | -----------------------------------------: |
| e         |       Scientific notation (lowercase e) |
| E         |       Scientific notation (uppercase E) |
| f         |                          Floating point |
| F         |                          Floating point |
| g         |                          General format |
| G         |   General format (uppercase scientific) |
| n         |              Locale-aware number format |
| \n        |                                New line |
| \         |   Escape character, Backslash (escaped) |
| :         |     Separates variable from format spec |
| .         |     Separates field width and precision |
| #         |                            Comment line |
| ()        |  Parentheses (grouping / function call) |
| ,         |             Separator between arguments |
| j         |       Imaginary unit in complex numbers |
| _         |  Underscore (allowed in variable names) |
| %         |                              Percentage |
| None      |                               Same as g |
