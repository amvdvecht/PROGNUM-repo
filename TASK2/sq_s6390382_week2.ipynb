# Checked. Question 2.12 : Quadrants

```python
from math import *

x = int(input("Give a value for x:"))
y = int(input("Give a value for y:"))
phi = atan2(y, x)
print(f"The angle between the line created from origin to (x,y) is: {degrees(phi)} degrees")

if y == 0 and x == 0 :
    print("The line to (x,y) is at the origin")
elif y > 0 and x > 0:
    print("The line to (x,y) is in the 1st quadrant")
elif y > 0 and x < 0:
    print("The line to (x,y) is in the 2nd quadrant")
elif y < 0 and x < 0:
    print("The line to (x,y) is in the 3rd quadrant")
elif y < 0 and x > 0:
    print("The line to (x,y) is in the 4th quadrant")
else:
    print("Yo, what happened?")
```
