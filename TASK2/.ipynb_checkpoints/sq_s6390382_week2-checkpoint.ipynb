# Checked. Question 2.12 : Quadrants

```python
from math import *

x = int(input("Give a value for x:"))
y = int(input("Give a value for y:"))
phi = atan2(y, x)
print(f"The angle between the line created from origin to (x,y) is: {degrees(phi)} degrees")

if y == 0 and x == 0 :
    print("The line to (x,y) is at the origin")
elif y > 0 and x > 0:
    print("The line to (x,y) is in the 1st quadrant")
elif y > 0 and x < 0:
    print("The line to (x,y) is in the 2nd quadrant")
elif y < 0 and x < 0:
    print("The line to (x,y) is in the 3rd quadrant")
elif y < 0 and x > 0:
    print("The line to (x,y) is in the 4th quadrant")
else:
    print("Yo, what happened?")
```

# Checked. Question 2.15 : Intersection of two lists

```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]
night1 = set(night1)
night2 = set(night2)

double = set.intersection(night1, night2)
double = list(double)
print(double)

if 5 in double:
    print("Number 5 is in both night1 and night2")
else:
    print("Number 5 is not in both sets")
if 5 in night1:
    print("Number 5 is in night1")
else:
    print("Number 5 is not nigh1")
```

# Question 2.21 : Fibonacci sequence

```python
# Define the starting values for the fibonacci sequence
first = 1
second = 2

for i in range(2, 100): # goes from the 3rd to 100th result because the 1st and 2nd are already defined and we want to go untill 100.
    following = first + second # Make next number in list the combination of the previous 2 numbers
    first = second # Make the second number the new first
    second = following # Replace the original second number with the calculated following

print(f"The 100th digit in the Fibonacci sequence is: {second}") #print the last second that was calculated, which is the 100th number
    
```
