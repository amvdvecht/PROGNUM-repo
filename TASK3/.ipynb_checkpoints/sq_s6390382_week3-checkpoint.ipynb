# Question 3.1 : Average calculator

```python
masses = [1.9891e+30, 1.8986e+27, 
          5.6846e+26, 1.0243e+26, 8.6810e+25,
          5.9736e+24, 4.8685e+24, 6.4185e+23, 
          3.3022e+23, 7.349e+22, 1.25e22]

smaller = []


for i in range(len(masses)):  # Makes a loop so the following code is done for indices ranging from 0 to the end of the list masses
    m = masses[i]    # Define new variable as specific item from masses with index i
    m_moon = masses[-2]   # Define the moon mass as the item with index -2
    
    if m <= m_moon:     # If the previously defined mass is smaller or equal to the moon mass do the following:
        smaller.append(m)   # Add the mass to the previously empty list "smaller"


print(smaller) # Print the list with only the smaller values


masses2 = masses[-5:]  # Define masses 2 as the list containing the last 5 of the original only
print(masses2)


avg = sum(masses2) / len(masses2) # Define the average as the sum of all values from masses2 devided by the length of masses2
print(avg)
```

# Question 3.2 : Calculate escape velocities

```python
from math import sqrt
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
v_esc = {} # Empty dictionary

for i in range(len(masses)): # Makes a loop so the following code is done for indices ranging from 0 to the end of the list masses
    M = masses[i]
    r = radii[i]
    name = names[i]   # Define mass radius and name for each index in lists
    v = sqrt((2*G*M)/r) # Calculate excape velocity
    print(f"The mass of the object is: {M:0.2e}. The escape velocity of the object is: {v:0.2e}")
    # or use v_esc[name] = v
    v_esc.update({name : v}) # Add item to empty dictionary
print(v_esc)
```

# Question 3.10 : Area and volume of a cylinder

```python
from math import *

def cylinder(r,h):
    A = 2 * pi * (r**2) + 2 * pi * r * h
    V = pi * (r**2) *h
    print(A, V)
    return(A, V)

A, V = cylinder(4,5)
ratio = A / V
print(ratio)
```

# Question 3.17 : Standard deviation using ndarray and in a one-liner

```python
def stdev(array):
    mean = np.mean(array) # Find mean of array with numpy
    sq_diff = (array-mean) **2 # Calculate squared difference
    sd_2 = np.mean(sq_diff) # Calculate mean of squared difference aka standard deviation squared
    return np.sqrt(sd_2) # Return square root of sd_2 which is the standard deviation

x = np.array([2.3, 3.5, 5.6, 12.8, 10.3, 7.2, 5.0, 2.1]) # Make a test array with numpy
print("With self defined function:", stdev(x))
print("With numpy function:", np.std(x))


# One-liner version, stdev_1
def stdev_1(array):
    return np.sqrt(np.mean((array-np.mean(array))**2))
print("With self defined one-line function:", stdev_1(x))
```

# Question 3.18 : Least Squares fitting

```python
import numpy as np # Import numpy with alias np

# Make numpy arrays out of both columns
# Distance (Mpc)
x = np.array([18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44.0, 66.7, 52.4, 55.0, 25.6, 26.19])
# Velocity (km/s)
y = np.array([2094.75, 2320.87, 1224.9, 872.13, 432.2, 751.0,1410.0, 2700.0, 5292.0, 2550.0, 5253.0, 2419.0, 2158.0])

# Calculate mean of x and y with numpy
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate a and b using the formulas given and numpy
b = np.sum((x - x_mean)*(y - y_mean)) / np.sum((x - x_mean)**2) # Works because np. automatically goes through each item seperately
a = y_mean - b*x_mean

print("Intercept a:", a,"km/s", "Slope/Hubble constant b:", b, "km/s/Mpc")

```

# Question 3.22 : Plotting a catenary

```python
import numpy as np # Import numpy with alias np
import matplotlib.pyplot as plt # Import a specific part of matplotlib with alias plt

# Define function catenary, cosh from the question, y in further calculations
def catenary(x):
    return (np.exp(x) + np.exp(-x)) / 2

# Plot function for integer x values between -5 and 6 with range()
x_range = list(range(-5, 6))
y_range = [catenary(x) for x in x_range]

# Plot the figure
plt.figure(figsize=(8, 6)) # Set figure size (width, height)

# Plot figure: x on x-axis, y on y-axis, markers show o at every x int, line color is blue, label makes a legend with y=cosh(x) in it
plt.plot(x_range, y_range, marker='o', color='blue', label='y = cosh(x)')

# Add labels to x and y axis both with fontsize 14
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Add extra details for clarity
plt.title('Catenary Curve', fontsize=16)  # Add title
plt.grid(True)                                          # Show grid
plt.legend(fontsize=12)                                 # Show legend

plt.show() # Shows plot
```
