| Command         |                                Description |
| :-------------- | -----------------------------------------: |
| print()         |                Prints output to the screen |
| input()         |           Gets user input (returns string) |
| type()          |             Returns the type of a variable |
| del             |                         Deletes a variable |
| import          |                           Imports a module |

| from … import * |      Imports all objects from module |
| keyword.kwlist  |                  List of reserved keywords |
| .value          |  Accesses the value attribute of an object |
| .real           |      Returns real part of a complex number |
| .imag           | Returns imaginary part of a complex number |

| List commands         |                                Description |
| :-------------- | -----------------------------------------: |
| a = [] | Makes an empty list |
| a = (a,b,c) | Makes a tuple (cant be edited)|
| a.append(b) | Adds b to the end of the list a |
|count(x)|Returns a count of the number of times the object x appears in the list|
|extend(x2)||Adds the items from list x2 to your list|
|insert(i,x)|Inserts object x before the element pointed to by index|
|index(x)|Returns the index for the first item in the list matching the object x|
|remove(x)| Deletes the first element from the list that matches the object x|
|sort()|Sort the list|
|copy()|Copy list to a new list (without sharing memory)|
|min(a) | Takes the lowest value from the list a|
|max(a) | Takes the highest value fromt the list a |
|sum(a) | Takes the sum of all values in a |
|len(a) | Takes the amount of items in list a |
| range(start, stop, step) | Take specific items from a list between a starting index and ending index with steps of a specific size |
|a[start:end:step]| slices start through not past end, by step|
|a[start:end]     | slices items from start through end-1|
|a[start:]        | slices items from start through the rest of the sequence|
|a[:end]         | slices items from the beginning through end-1|
|a[:]           | slices a copy of the whole sequence|
| a[N] | Takes only item with index N from the list a |

| Set commands         |                                Description |
| :-------------- | -----------------------------------------: |
intersection(A, B, C) Find common items in all 3 lists
A & B & C same as above
A.difference(B) find things that are NOT in both folders
