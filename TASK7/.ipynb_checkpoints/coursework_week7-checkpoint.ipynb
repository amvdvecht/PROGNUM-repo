### Simpson's rule (4 pt)  <font color='red'><b>COURSEWORK</b></font>

In section 'Simpson's rule' in the task documentation we plotted the function $f(x) = 1 + x + \sin(2x)$

* Create a object-oriented code that the input variables are a function you wish to integral in string and the integral intervals, `a` and `b`.
* Add a function in the object that calculates the integration with the Trapezium rule.
* Add a function in the object that calculates the integration with the Simpson’s rule.
* Add a function in the object that calculates the integration with Monte Carlo integration.
* Add a function in the object that calculates the integration with SymPy (``sympy.integrate()``).
* Add a function in the object that calculates the integration with SciPy (``quad()`` function).
* Use ``a=0.5`` and ``b=3.0`` as the test using all methods you create in the object. Print the test.
  
**Answers**:  

````python
import numpy as np
import sympy as sp
from scipy.integrate import quad
from sympy import sin, cos, exp, log # To make sure these math things work in the function

class Integrator:
    ''' This will calculate the integration in 5 different ways: Trapezium rule, Simpsons method, Monte Carlo integration, Sympy integration and scipy quad function'''
    def __init__(self, func_str, a, b, n=1000): # Set starting values and convert function into usable function
        #Define the initial function, a, b, and n
        self.func_str = func_str
        self.a = a
        self.b = b
        self.n = n if n % 2 == 0 else n + 1  # Simpson requires even n number of intervals to work
        #make sure x can work with sympy by making it a simpy symbol x
        self.x = sp.symbols('x')
        # Evaluate the function using simpy (aka. make it into a symbolic function)
        self.func_sym = eval(func_str, {"x": self.x, "sp": sp, "sin": sin, "cos": cos, "exp": exp, "log": log}) # Add all math things so simpy recognises them
        # Turn the symbolic function into a function that works with numpy
        self.func = sp.lambdify(self.x, self.func_sym, 'numpy')
        
    def trapezium(self): # integration using Trapezium rule
        x = np.linspace(self.a, self.b, self.n + 1) # Create n+1 values for x between a and b
        y = self.func(x) # y is the values of x from above substituted into the given function
        h = (self.b - self.a) / self.n # calculate the width of each interval
        
        return h * (0.5*y[0] + np.sum(y[1:-1]) + 0.5*y[-1]) #calculate the trapezium formula
        
    def simpson(self): # integration with the simpsons rule
        #Uses same first steps as trapezium integration
        x = np.linspace(self.a, self.b, self.n + 1)
        y = self.func(x)
        h = (self.b - self.a) / self.n

        return (h/3) * (y[0] + 4*np.sum(y[1:-1:2]) + 2*np.sum(y[2:-2:2]) + y[-1]) # Use the simpsons formula

    def monte_carlo(self, N=100000): # integration with monte carlo
        x_rand = np.random.uniform(self.a, self.b, N) # Create N random values for x between a and b
        y_rand = self.func(x_rand) # y is the function given evaluated for each x value
        
        return (self.b - self.a) * np.mean(y_rand) # calculate using monte carlo function

    def sympy_integral(self): # integration with sympy
        # use the simpy integration function on the function (sympy compatible form) with variable x between a and b
        result_sympy = sp.integrate(self.func_sym, (self.x, self.a, self.b)) 

        return float(result_sympy) # convert to float (sympy can give mathematical expressions instead of values so converting to float is necesarry to avoid this

    def scipy_quad(self): # integration with scipy (quad function)
        result_scipy, error = quad(self.func, self.a, self.b) # calculate the result using quad (add , error so the tuple created by the quad function is devided over result and error)

        return result_scipy
    
#Use given values to test the class:
func = '1+x+sin(2*x)' # HAS to be a string for the code to work
a = 0.5
b = 3.0
integrator = Integrator(func, a, b) # Input the function a and b into the class

# Pritn all options for the integration
print("Trapezium rule:", integrator.trapezium())
print("Simpsons rule:", integrator.simpson())
print("Monte Carlo:", integrator.monte_carlo())
print("Sympy integral:", integrator.sympy_integral())
print("Scipy quad:", integrator.scipy_quad())
````

### Fitting a Gaussian model for velocity profile data (16 pt) <font color='red'><b>COURSEWORK</b></font>

For this assignment we use the HI radio FITS cube
[ngc6946.fits](https://www.astro.rug.nl/intranet/courses/PROGNUMNEW/latest/DATA/ngc6946.fits)

It contains 101 channel maps with radio interferometer HI data.

1. (3 pt) A global HI profile is a set of measurements at fixed R.A. and Dec. as function of the spectral 
   axis (frequency or velocity). It shows at which frequency (or velocity) there is a peak in HI.
   Plot this profile for R.A. (x) index 58 and Dec. (y) index 55. 
   Label the X axis with the header value ``CUNIT3`` and the Y axis with the intensity 
   units in ``BUNIT``. For this data, the pixels represent brightness.
   
   In the global profile, the first pixel corresponds to 233 km/s and the last pixel corresponds to 377 km/s.
   Convert your X axis from pixel index numbers to km/s (use NumPy's ``linspace()`` to define the X values).
   
**Answer**:

````python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# Load FITS cube
filename = "ngc6946.fits"
hdul = fits.open(filename)
cube = hdul[0].data
header = hdul[0].header

# define given R.A and dec
ra_index = 58 # x index
dec_index = 55 # y index
profile = cube[:, dec_index, ra_index] # cube[z, y, x] (get all x for the fixed x and y, so RA and Dec from the cube)

# define maximum and minimum velocity as given in question (first pixel and last pixel)
v_min = 233   # km/s
v_max = 377   # km/s
n_channels = cube.shape[0]  # Number of channels in the cube

# Make the velocity axis by taking n_channels amount of points between the minimum and maximum velocity (converting pixel indices to km/s)
velocity = np.linspace(v_min, v_max, n_channels)

#Plot
plt.figure(figsize=(8,5))
plt.plot(velocity, profile, marker='*', markersize=3, linestyle='-')
#set labels to the correct units
plt.xlabel(header["CUNIT3"])
plt.ylabel(header["BUNIT"])

# Add title
plt.title("Global HI Profile at (RA=58, Dec=55)")
plt.grid(True) # add grid for clarity

plt.show()
````

2. (1 pt) Assume the profile can be modeled by a Gaussian. Make a function with signature:

          def f(x, amp, mu, sigma, zero):

   which returns the value of a Gaussian function with amplitude ``amp`` centered at
   position ``mu`` with width ``sigma`` and offset with respect to the x axis of ``zero``
   
**Answer**:   

````python
import numpy as np

def f(x, amp, mu, sigma, zero):
    return amp * np.exp(-((x - mu)**2) / (2 * sigma**2)) + zero # Fill in gaussian function from notebook
````

**Answer**:

however, mu and sigma are both squared so their partial derivatives will depend on them thus making the whole function not linear in all parameters, thus not linear at all.

Initial estimates are essential for a good fit. If they are not close enough to the real values, you can find a fit but without a guarantee that it is the best fit.

4. (3 pt) For a least squares fit with function `scipy.optimize.curve_fit()` we need reasonable initial estimates. Find reasonable estimates and motivate why you think they are reasonable and then use them as argument ``p0`` in ``curve_fit()`` to find the best fit model parameters and their (scaled) standard errors    

**Answer**:

````python
from scipy.optimize import curve_fit
import numpy as np

# estimates
zero0 = np.min(profile) # Take the minimum value to be the baseline
amp0 = np.max(profile) - zero0 # the amplitude is the maximum value minus the baseline
mu0 = velocity[np.argmax(profile)] # mu is the value from the velocities that corresponds to the index of the maximum height

# estimate sigma using FWHM = 2 * sqrt(2 * ln(2)) * sigma
half_max = zero0 + (amp0 / 2) # Calculate half of the maximum amplitude
indices = np.where(profile > half_max)[0] # Get the indices of all values above the half maximum to use for the velocities again
sigma0 = (velocity[indices[-1]] - velocity[indices[0]]) / (2 * np.sqrt(2 * np.log(2)))

# Argument with initial estimates
p0 = [amp0, mu0, sigma0, zero0]

# use curve fit using funciton f, the velocity, profile and initial estimates p0.
#[0] = best amplitude
#[1] = best center (mu)
#[2] = best width (sigma)
#[3] = best offset
param, uncert = curve_fit(f, velocity, profile, p0=p0)

# best-fit parameters
amp_fit = param[0]
mu_fit = param[1]
sigma_fit = param[2]
zero_fit = param[3]

# uncertainties written out explicitly (no np.diag)
amp_err = np.sqrt(uncert[0, 0])
mu_err = np.sqrt(uncert[1, 1])
sigma_err = np.sqrt(uncert[2, 2])
zero_err = np.sqrt(uncert[3, 3])

print("Best-fit parameters:")
print(f"amp = {amp_fit:.4f} +- {amp_err:.4f}")
print(f"mu = {mu_fit:.4f} +- {mu_err:.4f}")
print(f"sigma = {sigma_fit:.4f} +- {sigma_err:.4f}")
print(f"zero = {zero_fit:.4f} +- {zero_err:.4f}")
````

Fitting multiple signals with a Gaussian.

6. (3 pt) If you fit is alright, you see that a small signal at the left of the main signal in your
   profile has not been fitted. You can add a second Gaussian (without a zero level because you 
   already have one offset) to model that second signal.
   Find the best fit parameter of this composed model and plot original data with the best fit curve 
   
**Answer**:

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# make new gaussian model function for two gaussian shapes in one
def f2(x, amp1, mu1, sigma1, amp2, mu2, sigma2, zero):
    g1 = amp1 * np.exp(-((x - mu1)**2) / (2 * sigma1**2))
    g2 = amp2 * np.exp(-((x - mu2)**2) / (2 * sigma2**2))
    return g1 + g2 + zero

# initial values from previous estimates (for the big gaussian peak)
amp1_0 = amp_fit
mu1_0 = mu_fit
sigma1_0 = sigma_fit
zero0 = zero_fit

# initial values (for the small peak to the left of the big peak)
left_mask = velocity < 290 #only look at values fully below the big peak (so it will actually take the max of the smaller peak)
amp2_0 = np.max(profile[left_mask]) - zero0 # calculate the amplitude in the same way as for the bigger but now excluding anythign including and after the big peak
mu2_0 = velocity[left_mask][np.argmax(profile[left_mask])] # value of velocity corresponding to new maximum height
sigma2_0 = 5  # peak has a width of 10-12 so sigma of 5 is good estimate

p0 = [amp1_0, mu1_0, sigma1_0, amp2_0, mu2_0, sigma2_0, zero0]

# apply to curve fit
param2, uncert2 = curve_fit(f2, velocity, profile, p0=p0)

# use the function f2
model2 = f2(velocity, *param2)

# plot the data and the new model with 2 gaussians
plt.figure(figsize=(8,5))
plt.plot(velocity, profile, '*', markersize=3, label="Data")
plt.plot(velocity, model2, '-', linewidth=2, label="Double-Gaussian fit")

plt.xlabel(header["CUNIT3"])
plt.ylabel(header["BUNIT"])
plt.title("HI Profile: Double Gaussian Fit")
plt.legend()
plt.grid(True)
plt.show()
````
